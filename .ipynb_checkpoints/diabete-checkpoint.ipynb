## 3. Treatment of outliers

FInd if the dataset contains the nan, null or duplicated values .

```python
data.isnull().sum()
```

```python
data.isna().sum()
```

```python
data.duplicated()
```

```python
data.shape
```

```python
pd.DataFrame(data=[data[columns].skew(),data[columns].kurtosis()],index=['skewness','kurtosis'])
```

```python
from scipy.stats import zscore
# for each column , we print the number of outlier
for i in columns:
    y_outliers = data[abs(zscore(data[i])) >= 3 ]
    print(f"{i} has {len(y_outliers)} outliers")
    y_outliers
```

We can remove the outliers.

```python
for i in columns:
    q1 = data[i].quantile(0.25)
    q3 = data[i].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    data = data[(data[i] > lower_bound) & (data[i] < upper_bound)]
```

Then let's us see the the value of Skewness and kurtosis of each column

```python
pd.DataFrame(data=[data[columns].skew(),data[columns].kurtosis()],index=['skewness','kurtosis'])
```

We create another column to explain if a person is diabetic.

```python
def outcome_class(Outcome):
    if Outcome ==1:
        return "Diabetic"
    else:
        return "No diabetic"

data['Classe'] = data['Outcome'].apply(outcome_class)
data.sample(frac=1).head()
```

## 3. Model Training and Test



```python
# Create X variable that contains features without the target __Outcome__. 
X = data.drop(['Classe','Outcome'],axis=1)
X
```

```python
# give the target to Y variable
Y = data['Outcome']
Y
```

```python
# training of model
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
# features normalization
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

We use different to find the best for ou project. 

### 3.1 Logistics Regression.



```python
classifier_lr = LogisticRegression(random_state = 0)
classifier_lr.fit(X_train, Y_train)
```

```python
Y_pred = classifier_lr.predict(X_test)
```

```python
accuracy_logistic = accuracy_score(Y_test, Y_pred)
print(f"The accuracy is  {accuracy_logistic}:")
```

```python
predictions = classifier_lr.predict(X_test)
print("The error  is :")
mean_squared_error(predictions, Y_test)
```

```python
# Matrix confusion is used to determine the number of True-Positive, False-Postive, True-Negative and False-Negative
confusion_matrix = metrics.confusion_matrix(Y_test, Y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.grid(False)
plt.show() 
```

## 3.3 Decision Tree Algorithm


```python
classifier_dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier_dt.fit(X_train, Y_train)
```

```python
Y_pred = classifier_dt.predict(X_test)
```

```python
accuracy_dt = accuracy_score(Y_test, Y_pred)
print(f"The accuracy is  {accuracy_dt}:")
```

```python
from sklearn.metrics import mean_squared_error
predictions = classifier_dt.predict(X_test)
print("L'erreur  est :")
mean_squared_error(predictions, Y_test)
```

```python
confusion_matrix = metrics.confusion_matrix(Y_test, Y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.grid(False)
plt.show() 
```

```python

vn, fp, fn, vp = confusion_matrix.ravel()
 
print("True Positive (TP): ", vp)
print("True Negative (TN): ", vn)
print("False Positive (FP): ", fp)
print("False Negative (FN): ", fn)
 
# Calcul de accuracy
accuracy = (vp + vn) / (vp + vn + fp + fn)
 
# Calcul de recall
recall = vp / (vp + fn)
 
# Calcul de F1-score
 
print("\n\nAlgorithm:")
print("Accuracy: (VP + VN) / (VP + VN + FP + FN)")
print("Recall: VP / (VP + FN)") 
print("\n\nRsults:")
print("Accuracy: ", round(accuracy, 14))
print("Recall: ", round(recall, 14))
```

```python
# test
print("The  predicted result is")
print(classifier_kn.predict([[2.0,1.0,12.0,1.0,0.0,6.6,3.27,5.0]]),"\n")
print("The actual result is :")
print(data['Outcome'][0])
```
